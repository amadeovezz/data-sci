# Problem statement and some Exploratory Data Analysis

You are employed by the UACA (United Association of Crocodiles and Aligators) and have a new task for you. Some of your colleagues,
(data collection officers) travel around the world and track crocodiles in the wild. They have just gotten back from a recent trip and
have a problem. A new data collection officer Antonio, has an incredibly hard time distinguishing between alligators and crocodiles.
Thus, the senior team members are curious if you can build something that helps Antonio differentiate between the two species.

Note: This data is NOT real data. However, it does try to encode known heuristics (given below) - so it bares
some similarity as to what it might be in the real world.

## Features

- weight (pounds): real number
- num_of_teeth: This feature is the observed number of teeth by the data collection officers. This is a semi reliable feature given that they are counting teeth from a distance and the reptiles are sometimes only observable in water. represented as an integer
- length (feet): real number
- angle_from_snout (degrees): The angle from the tip of the snoot to the eyes of a crocodile. Positive real number
- location (country):
- species: 'character'

## Known heuristics:

- Crocodiles are longer than alligators
- Crocodiles have more visible teeth than alligators
- Crocodiles have pointier snouts
- Alligators are heavier than crocodiles when they are the same length
- Alligators are only found in the US, China, and a very small population exists in Mexico, whereas crocodiles are found all over the world

Before we start our analysis it is helpful to visualize our data:

```python
import pandas as pd
import seaborn as sns

# Import pseudo data set
df = pd.read_csv('../data/tabular/croc-aligator.csv')
training_data = pd.DataFrame(
    {'weight': df.weight
    ,'length': df.length
    , 'num_of_teeth': df.num_of_teeth
    , 'angle_from_snout': df.angle_from_snout
    , 'location': df.location}
).to_numpy()

labels =  pd.DataFrame({'species':df.species}).to_numpy() # age
#Visualize the data
sns.scatterplot(data=df, x="weight", y='length', hue='species', legend='full')
```
